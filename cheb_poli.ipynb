```octave
x=linspace(-1.005,1.005,1000);
y=512*x.^10 - 1280*x.^8 + 1120*x.^6 - 400*x.^4 + 50*x.^2 - 1;
plot(x,y)
hold on;
plot([-1.5 1.5], [0 0], 'k', 'LineWidth', 1);
plot([0 0], [-2 2], 'p', 'LineWidth', 1);
title("The Chebyshev polinomial of degree 10")
```
