```octave
%%RKC method,s=5
[X,Y] = meshgrid(linspace(-20,1), linspace(-8,8));
Z = X+Y*1i;
phi = 51/75+24/75*(16 *(1+Z/8).^5-20*(1+Z/8).^3+5*(1+Z/8))
contourf(X,Y,1-abs(phi), [0 0], 'LineWidth', 1);
set(gca,'FontSize', 10, 'CLim', [0 1]);
colormap([.1 .5 .3; 0 0 0; 1 1 1]);
hold on;
plot([-20 1], [0 0], '--k', 'LineWidth', 1);
plot([0 0], [-8 8], '--k', 'LineWidth', 1);
axis('square');
title('Absolute stability region of the second order RKC method, s=5')
```
