```octave
function T=Chebyshev5(x)
T=16 *(x).^5-20*(x).^3+5*(x);
end
function T=Chebyshev5der(x)
T=80 *(x).^4-60*(x).^2+5;
end
```

```octave
%%RKC method,s=5
[X,Y] = meshgrid(linspace(-60,1), linspace(-8,8));
Z = X+Y*1i;
epsilon=0.05;
w_0=1+epsilon/25;
w_1=Chebyshev5(w_0)/Chebyshev5der(w_0);
phi = Chebyshev5(w_0+w_1*Z)/ Chebyshev5(w_0);;
contourf(X,Y,1-abs(phi), [0 0], 'LineWidth', 1);
set(gca,'FontSize', 10, 'CLim', [0 1]);
colormap([.1 .5 .3; 0 0 0; 1 1 1]);
hold on;
plot([-60 1], [0 0], '--k', 'LineWidth', 1);
plot([0 0], [-8 8], '--k', 'LineWidth', 1);
axis('square');
title('Absolute stability region of the first order damped RKC method, s=5, epsilon=0.05')
```

```octave
function T=Chebyshev10(x)
T=512*x.^10 - 1280*x.^8 + 1120*x.^6 - 400*x.^4 + 50*x.^2 - 1;
end
function T=Chebyshev10der(x)
T=512*10*x.^9-1280*8*x.^7+1120*6*x.^5-400*4*x.^3+50*2*x;
end
```

```octave
%%RKC method,s=10
[X,Y] = meshgrid(linspace(-200,1), linspace(-24,24));
Z = X+Y*1i;
epsilon=0.02;
w_0=1+epsilon/25;
w_1=Chebyshev10(w_0)/Chebyshev10der(w_0);
phi = Chebyshev10(w_0+w_1*Z)/ Chebyshev10(w_0);
contourf(X,Y,1-abs(phi), [0 0], 'LineWidth', 1);
set(gca,'FontSize', 10, 'CLim', [0 1]);
colormap([.1 .5 .3; 0 0 0; 1 1 1]);
hold on;
plot([-200 1], [0 0], '--k', 'LineWidth', 1);
plot([0 0], [-24 24], '--k', 'LineWidth', 1);
axis('square');
title('Absolute stability region of the first order damped RKC method, s=10, epsilon=0.01')
```
